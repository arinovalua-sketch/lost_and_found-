```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
scores = [85, 92, 78, 90, 88]
```

```python
plt.plot(names, scores)
plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Student Scores Line Plot')
plt.show()
```

```python
plt.bar(names, scores)
plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Student Scores Bar Chart')
plt.show()
```
